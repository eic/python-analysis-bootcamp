# Accessing Truth Information

An important aspect at this stage in the design of EIC experiments is the comparison of the so-called 'truth' information with simulated variables. In this notebook we will go over how to access the truth information in an ATHENA event.

## Importing uproot

Depending on the versions of uproot and XRootD that you have installed, you may encouter a warning from uproot below. Nevertheless, because of the simple data format of the ATHENA ROOT files, we are able to ignore this warning.

```python
import uproot as ur
print('Uproot version: ' + ur.__version__)
```

## Opening a file with uproot

To test uproot, we will open a sample file (a single-particle simulation of interest to those who wish to study detector performance):

```python
server = 'root://sci-xrootd.jlab.org//osgpool/eic/'
file = 'ATHENA/RECO/master/SINGLE/pi+/1GeV/3to50deg/pi+_1GeV_3to50deg.0001.root'
```

```python
events = ur.open(server + file + ':events', library = 'np')
```

## Truth information in the `mcparticles2` branch

For some historical reasons which are unlikely to change in the near future, truth information is stored in the `mcparticles2` branch. This includes all *steps* in the simulation, subject to certain conditions. For the purpose of end-user analysis, the conditions are essentially that only steps by primary particles are included.

Several fields are available for the truth information:

```python
events['mcparticles2'].keys()
```

Besides the particle data group code `pdgID`, the parent track `g4Parent`, and the generator status `genstatus`, you will also see the vertex (`v`) and momentum (`p`) of at the start (`s`) and end (`e`) of each step. Thus, `ps.x` corresponds to the `x` component of the momentum at the start of a step. Let's retrieve these starting momenta, as well as the `pdgID` and `g4Parent` code

```python
pdgID = events['mcparticles2.pdgID'].array()
g4Parent = events['mcparticles2.g4Parent'].array()
genStatus = events['mcparticles2.genStatus'].array()
psx,psy,psz = events['mcparticles2.ps.x'].array(), events['mcparticles2.ps.y'].array(), events['mcparticles2.ps.z'].array()
```

As expected, for this file the `pdgID` corresponds to that of a pion.

```python
pdgID[100]
```

```python
print(g4Parent[100])
```

And indeed, the energy of this pion is 1 GeV. We are importing the `numpy` library to use the `sqrt` function.

```python
import numpy as np
p = np.sqrt(psx**2 + psy**2 + psz**2)
```

```python
p[100]
```

## Making a simple plot

We can now create a simple plot of the angular (theta) distribution of the generated particles.

```python
import awkward as ak
theta = np.arctan2(np.sqrt(psx**2 + psy**2), psz)
ak.flatten(theta[g4Parent == 0])
```

```python
import matplotlib.pyplot as plt
plt.hist(ak.flatten(theta[g4Parent == 0]), bins = 50)
plt.xlabel('Initial Scattering Angle $\\theta$ [rad]')
plt.ylabel('Number of events')
plt.show()
```
